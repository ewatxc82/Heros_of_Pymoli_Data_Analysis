```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
df = pd.read_csv(file_to_load)
df.head()
```

## Player Count

* Display the total number of players


```python
num_players = df['Gender'].count()
print("Total number of players is " + str(num_players))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = df['Item ID'].unique()
num_unique_item = len(unique_item)
avg_price = df['Price'].mean()
purchases = len(df['Purchase ID'])
total_revenue = sum(df['Price'])
summary_df = pd.DataFrame({"Total Unique Items": [num_unique_item],
                          "Average Price": [avg_price],
                          "Total Purchases": [purchases],
                          "Total Revenue": [total_revenue]})
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_df = df['Gender'].value_counts()
percent_male = gender_df['Male'] / num_players 
percent_female = gender_df['Female'] / num_players
percent_other = gender_df['Other / Non-Disclosed'] / num_players
gender_dem_df = pd.DataFrame({
                "Percentage of Male Players": [percent_male],
                "Count of Male Players": [gender_df['Male']],
                "Percentage of Female Players": [percent_female],
                "Count of Female Players": [gender_df['Female']],
                "Percentage of Other / Non-Disclosed Players": [percent_other],
                "Count of Other / Non-Disclosed Players": [gender_df['Other / Non-Disclosed']]})
gender_dem_df.head()
```
